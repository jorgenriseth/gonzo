# FreeSurfer vs FastSurfer
FreeSurfer and/or FastSurfer are used for image segmentation and surface reconstructions in the Gonzo processing-pipeline.
FastSurfer is a deep-learning based drop-in alternative for FreeSurfer.
Brain segmentations are used in several parts of the processing pipeline for computing regionwise statistics of different MRI-sequences, masking parts of the brain in computations or for generating contour surfaces of e.g. ventricles.
The computational mesh generation relies on 3D surfaces either output from FreeSurfer 

If you have no preference for using FreeSurfer or FastSurfer, I recommend FastSurfer.
FastSurfer is usually faster, and, for the single subject included in this dataset, the outputted pial- and white-matter surfaces have less artifacts causing issues for the meshing algorithm.

A FreeSurfer-license is still required for FastSurfers surface reconstruction pipeline.

## Running FastSurfer

We have created a wrapper script for running FastSurfer in a singularity container, so the user do not have to install it on their own computer. 

```bash
python scripts/fastsurfer.py \
    -t1 mri_dataset/sub-01/ses-01/anat/sub-01_ses-01_T1w.nii.gz \
    -o mri_processed_data/fastsurfer/sub-01 \
    -l docker/license.txt  \
    --threads 12 \
```

## Running FreeSurfer

Similarly, we created a wrapper script for running FreeSurfer in a singularity container.
